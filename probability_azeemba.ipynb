```python
import numpy as np
import matplotlib.pyplot as plt

def expected_value(x, outcomes):
    expected = 0
    for i in range(len(outcomes)):
        expected += x[i]*outcomes[i]

    expected = expected/sum(outcomes)
    return expected

def plot_distribution(x, counts, show=True, **kwargs):
    total = sum(counts)
    percentages = [(count / total) * 100 for count in counts]
    
    # Plotting the histogram using bar plot
    plt.bar(x, percentages, width=1, edgecolor='black', **kwargs)
    
    # Adding titles and labels
    plt.title('Histogram')
    plt.xlabel('Result')
    plt.ylabel('Percentage')
    
    # Display the plot
    if show:
        plt.show()
    return plt
```

```python
# D20 advantage
results = [0]*20
for i in range(0, 20):
    for j in range(0, 20):
        results[max(i, j)] += 1

results
```

```python
# D20 advantage expected value
expected = expected_value(list(range(1, 21)),results)
expected
```

```python

# Define the bin edges (one more than the number of results)
x_values = list(range(1, 21))
plot_distribution(x_values, results)
```

```python
# D6
res = [0]*12

for i in range(6):
    for j in range(6):
        res[i+j+1] += 1

res
```

```python
sum(res)
x_2d6 = list(range(1, 13))
expected_2d6 = expected_value(x_2d6,res)
expected_2d6
```

```python
plot_distribution(x_2d6, res)
```

```python
# D12
r_12 = [1]*12
x_12 = list(range(1,13))
expected_d12 = expected_value(x_12, r_12)
plot_distribution(x_12, r_12)
```

```python
plt.figure(figsize=(720*2/100, 400*2/100))
plt.rcParams.update({'font.size':24})
plot_distribution(x_2d6, res, show=False, label="6-sided dice (2d6)")
plt = plot_distribution(x_12, r_12, show=False, alpha=0.3, label="12-sided dice (1d12)")

plt.title("Outcome Probabilities")
plt.legend()
plt.savefig("probabilities-1d12-vs-2d6.png")
plt.show()
```
